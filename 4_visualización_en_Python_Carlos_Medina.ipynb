<h2>Consigna 3: Realizar un gráfico de violín con los salarios, discriminados por género</h2>

```python
import seaborn as sns
```

Hay errores. los 'yes' y los 'no' algunos tienen diferencias en minusculas y mayusculas

```python
df_recursosHumanos['HispanicLatino']=df_recursosHumanos['HispanicLatino'].astype(str)
df_recursosHumanos['HispanicLatino']=df_recursosHumanos.HispanicLatino.str.replace('no','No')
df_recursosHumanos['HispanicLatino']=df_recursosHumanos.HispanicLatino.str.replace('yes','Yes')
df_recursosHumanos
```

Tome el campo de latino para completar la data con tres parametros y que la grafica violin quedase bien

```python
salaryGender = sns.catplot(data=df_recursosHumanos, 
                     kind='violin', 
                     x='HispanicLatino',
                     y='Salary',
                     hue='GenderID',
                     split=True
                     )

salaryGender.set(xlabel='Latino', 
       ylabel='Salario', 
       title='Salarios por genero y raza')
```

<h2>Consigna 4: Graficar la serie de tiempo correspondiente a la fecha de contratación (DateofHire)</h2> 

Pasar a formato fecha

```python
pd.to_datetime(df_recursosHumanos.DateofHire)
pd.to_datetime(df_recursosHumanos.DateofTermination)
```

Extraer el año en una nueva columna

```python
df_recursosHumanos['yearHire']=pd.DatetimeIndex(df_recursosHumanos['DateofHire']).year
print(df_recursosHumanos['yearHire'])
df_recursosHumanos['yearTermination']=pd.DatetimeIndex(df_recursosHumanos['DateofTermination']).year
df_recursosHumanos['yearTermination']
```

Contar cada dato por año

```python
df_countYearHire=df_recursosHumanos.groupby(df_recursosHumanos.yearHire).yearHire.count()
df_countYearHire
```

```python
df_countYearTermination=df_recursosHumanos.groupby(df_recursosHumanos.yearTermination).yearTermination.count()
df_countYearTermination
```

Graficar juntos

```python
fig2, ax2 = plt.subplots(figsize=(12, 3))
ax2.plot(df_countYearHire.index, df_countYearHire, Label='Contrataciones en el tiempo', color='blue')
ax2.plot(df_countYearTermination.index, df_countYearTermination, Label='Terminaciones en el tiempo', color='red')
ax2.set_xlabel('Año')  				
ax2.set_ylabel('Contrataciones')
ax2.set_title('Contrataciones y terminaciones en el tiempo')
ax2.legend()
```
